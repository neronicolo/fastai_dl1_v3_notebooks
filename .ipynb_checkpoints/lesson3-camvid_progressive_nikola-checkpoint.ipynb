# Stage 1


## Data Setup and Preview

```python
path = untar_data(URLs.CAMVID, dest=Path().home()/'programming/data')
path.ls() 
```

```python
path_imgs = path/'images'
path_lbls = path/'labels'
```

```python
img_files = get_files(path_imgs)
img_files[:5]
```

```python
lbl_files = get_files(path_lbls)
lbl_files[:5]
```

```python
img_file = img_files[0]
img = open_image(img_file)
img.show(figsize=(6,6))
```

```python
get_y_fname = lambda x: path_lbls/f'{x.stem}_P{x.suffix}'
get_y_fname(img_file)
```

```python
img_mask = open_mask(get_y_fname(img_file))
img_mask.show(figsize=(6,6))
```

```python
img_size = np.array(img_mask.size)
img_size, img_mask.data
```

```python
codes = np.loadtxt(path/'codes.txt', dtype=str)
codes
```

```python
codes[[4,21,22]]
```

```python
size = img_size//2
xform = get_transforms(max_lighting=0.4, max_warp=0.1)

free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=8
else:           bs=2
print(f"img_size={size},using bs={bs}, have {free}MB of GPU RAM free")
```

```python
src = (SegmentationItemList.from_folder(path_imgs)
      .split_by_fname_file(path/'valid.txt')
      .label_from_func(get_y_fname, classes=codes))
```

```python
data = (src.transform(tfms=xform, size=size, tfm_y=True)
       .databunch(bs=bs)
       .normalize(imagenet_stats))
```

```python
data.show_batch(rows=2, figsize=(10, 12))
```

```python
data.show_batch(rows=2, figsize=(10, 12), ds_type=DatasetType.Valid)
```

## Training

```python
codes
```

```python
name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['Void']

def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
metrics = acc_camvid
# metrics=accuracy
```

```python
wd = 1e-2
```

```python
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(3e-03))
```

```python
learn.save('stage_1')
```

```python
learn.load('stage_1');
```

```python
learn.show_results(rows=2, figsize=(9,9))
```

## Fine Tuning

```python
learn.load('stage_1');
```

```python
learn.unfreeze()
learn.lr_find(start_lr=1e-09, end_lr=1e-02)
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-06, 1e-04))
```

```python
learn.save('stage_1_fine_tuning')
```

```python
learn.load('stage_1_fine_tuning');
```

```python
learn.show_results(rows=2, figsize=(9,9))
```

# Stage 2

## Data Setup

```python
#learn.destroy()

size = img_size
xform = get_transforms(max_lighting=0.4, max_warp=0.1)

free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=3
else:           bs=1
print(f"img_size={size},using bs={bs}, have {free}MB of GPU RAM free")
```

```python
data = (src.transform(tfms=xform, size=size, tfm_y=True)
       .databunch(bs=bs)
       .normalize(imagenet_stats))
```

```python
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
learn.load('stage_1_fine_tuning');
learn = learn.to_fp16()
learn.freeze()
learn.data = data
learn.data
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-04))
```

```python
learn.save('stage_2')
```

## Fine Tuning

```python
learn.load('stage_2');
```

```python
learn.unfreeze()
learn = learn.to_fp16()
```

```python
learn.lr_find(start_lr=1e-08, end_lr=1e-02)
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, slice(5e-06, 5e-04))
```

```python
learn.save('stage_2_fine_tuning')
```

```python
learn.show_results(rows=2, figsize=(9,9))
```

```python
learn.export(file='camvid_progressive_nikola.pkl')
```
