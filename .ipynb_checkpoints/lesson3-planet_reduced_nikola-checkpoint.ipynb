## Preparing data for training - data_block API/DataBunch()

```python
df = pd.read_csv(path/'train_v2.csv')
df.head()
```

```python
xform = get_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0.)
```

```python
np.random.seed(42)
src = (ImageList.from_csv(path, 'train_v2.csv', folder='train-jpg', suffix='.jpg')
      .split_by_rand_pct()
      .label_from_df(label_delim=' '))
```

```python
data = (src.transform(xform, size=128)
       .databunch()
       .normalize(imagenet_stats))
```

```python
data.show_batch(rows=3, figsize=(12,9))
```

```python
data
```

## Traininig

```python
arch = models.resnet50
```

```python
acc_02 = partial(accuracy_thresh, thresh=0.2)
f_score = partial(fbeta, thresh=0.2)
learn = cnn_learner(data, arch, metrics=[acc_02, f_score])
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 1e-02
```

```python
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-rn50')
```

#### Fine tuning

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, slice(1e-05, lr/5))
```

```python
learn.save('stage-2-rn50')
```

## Progressive training 

Increasing image size and re-training model with bigger image resolution.

```python
learn.load('stage-2-rn50');
```

```python
data = (src.transform(xform, size=256)
        .databunch(bs=32)
        .normalize(imagenet_stats))
```

```python
learn.data = data
```

```python
data.train_ds[0][0].shape
```

```python
## train just head of the network with new image resolution
learn.freeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lr = 1e-3
```

```python
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-256-rn50')
```

```python
learn.load('stage-1-256-rn50');
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, slice(1e-05, lr/5))
```

```python
learn.save('stage-2-256-rn50')
```

```python
learn.recorder.plot_losses()
```
